```python
from sklearn.preprocessing import MinMaxScaler
L=MinMaxScaler()
df['Time']=L.fit_transform(df[['Time']])
df['Netflow_Bytes']=L.fit_transform(df[['Netflow_Bytes']])
```

```python
from sklearn.preprocessing import LabelEncoder
M=LabelEncoder()
df['Protcol']=M.fit_transform(df['Protcol'])
df['Flag']=M.fit_transform(df['Flag'])
df['Family']=M.fit_transform(df['Family'])
df['SeddAddress']=M.fit_transform(df['SeddAddress'])
df['ExpAddress']=M.fit_transform(df['ExpAddress'])
df['IPaddress']=M.fit_transform(df['IPaddress'])
df['Threats']=M.fit_transform(df['Threats'])
```

```python
from sklearn.model_selection import train_test_split
X=df.iloc[:,0:13]
Y=df.iloc[:,13]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
X.info()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
model=GaussianNB()
model.fit(X_train,Y_train)
print(accuracy_score(Y_test,model.predict(X_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(X_train, Y_train)
Y_pred = rfc.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print(accuracy)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svc = SVC(kernel='linear')
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print(accuracy)
```
